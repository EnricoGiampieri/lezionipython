##Introduzione a Pylab##

Pylab è una superlibreria che include tre piloni del calcolo scientifico in python:

* **numpy**, che gestisce vettori e matrici in modo naturale
* **scipy**, che definisce un gran numero di algoritmi su questi vettori
* **matplotlib**, una libreria di plotting dei risultati

Sia numpy che matplotlib sono stati creati a partire da una sintassi simile (ma non identica) a quella di MATLAB, 
per cui i suoi utenti dovrebbero ritrovarsi abbastanza a proprio agio, dovendo però sempre ricordare che in python
gli indici partono da 0 e non da 1.

```python
import numpy as np
import scipy as sp
import pylab as pl
#figsize(12,6)
```

Sono definiti tutti gli operatori matematici fra array, che effettua le operazioni su tutti gli elementi.

```python
a = np.r_[0:10]
b = np.r_[0:10]
print a*b
print a+b
print a.dot(b)
```

Gli array possono contenere qualsiasi tipo di dati, fintanto che tutti gli elementi sono dello stesso tipo.

```python
a = array([ 1.+i*1j for i in range(5) ])
print a
print a+a
print a*a
print a**a
```

Non è detto però che le operazioni matematiche abbiano senso su ogni insieme di elementi!

```python
a = array(["pippo","pluto","tizio","caio"])
print a
print a+a
```

Le operazioni algebriche di base non sono possibili fra array con un numero di dimensioni diverse.
Numpy è però in grado di ripetere automaticamente un array lungo una dimensione trasversale per renderlo
di dimensioni compatibili con quelle dell'altro array. Questa operazione è chiamata *broadcasting*.
Per ottenere l'array trasposto possiamo ottenerlo tramite l'attributo T dell'array.

```python
a = np.array(xrange(4))
b = np.zeros((4,4))
c = a+b
print a
print b
print c
print c.T
```

##Integrazione di equazioni differenziali

```python
from scipy.integrate import odeint
tempi = sp.r_[0.:10.:101j]
def campo(s,t):
    x,y = s
    return -y,x
res=odeint(campo, (1,0), tempi)
x,y = res.T[:]
plot(tempi,x)
plot(tempi,y);
```

```python
t = numpy.linspace(0, 4*numpy.pi, 20)
cos = numpy.cos
sin = numpy.sin

x = sin(2*t)
y = cos(t)
z = cos(2*t)
s = 2+sin(t)

mlab.points3d(x, y, z, s, colormap="copper", scale_factor=.25)
mlab.show()
```

```python
n_mer, n_long = 6, 11
pi = numpy.pi
dphi = pi/1000.0
phi = numpy.arange(0.0, 2*pi + 0.5*dphi, dphi)
mu = phi*n_mer
x = numpy.cos(mu)*(1+numpy.cos(n_long*mu/n_mer)*0.5)
y = numpy.sin(mu)*(1+numpy.cos(n_long*mu/n_mer)*0.5)
z = numpy.sin(n_long*mu/n_mer)*0.5

mlab.plot3d(x, y, z, numpy.sin(mu), tube_radius=0.025, colormap='Spectral')
mlab.show()
```

```python
from scipy.integrate import odeint
import scipy as sp
def lorenz(s,t):
    x,y,z = s
    dx = 10.*(y-x)
    dy = x*(28.-z)-y
    dz = x*y - 8./3*z
    return dx,dy,dz

tempi = sp.r_[0.:100.:0.01]
res=odeint(lorenz, (0.,1.,1.15), tempi)
x,y,z = res.T[:]
mlab.plot3d(x, y, z, tempi, tube_radius=0.125, colormap='Spectral')
mlab.show()
```
