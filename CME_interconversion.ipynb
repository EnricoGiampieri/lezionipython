```python
%load_ext sympy.interactive.ipythonprinting
from CMElib_new import CME
from sympy import Symbol
import math
```

```python
A = Symbol('A')
B = Symbol('B')
```

```python
# interconversion
cme = CME()
cme.add_reaction(A, B, 0.1*A)
cme.add_reaction(B, A, 0.1*B)
```

```python
#visualizing the  CME
display(cme.writeCME())
```

```python
time, functions_dict = cme.evaluate({A:10000, B:0}, 100, A, B, A+B)
for f_name, f_values in functions_dict.items():
    pylab.plot(time, f_values, label=f_name, linestyle='steps-mid')
    
#setting the margins
pylab.gca().margins(0.0,0.1)
pylab.legend()
    
```

```python
# stationary distribution
# with  gillespie
distributions = cme.distribution({A:50, B:0},5000,500, A, B)
pylab.plot(distributions[A], linestyle='steps-mid', label='A')
pylab.plot(distributions[B], linestyle='steps-mid', label='B')
pylab.legend()
pylab.show()
```
