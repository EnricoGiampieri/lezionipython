```python
%load_ext sympy.interactive.ipythonprinting
from CMElib_new import CME
from sympy import Symbol
import math
```

```python
#degradation reaction
A= Symbol('A')
cme = CME()
cme.add_reaction(A,None, .1*A)
```

```python
#bacterial growth
cme = CME()
cme.add_reaction(None, A, .1*A)
```

```python
#logistic growth
A= Symbol('A')
cme.add_reaction(A, None, A**2)
cme.add_reaction(None, A, A)
```

```python
# interconversione
cme = CME()
cme.add_reaction(A, B, A/10)
cme.add_reaction(B, A, B/10)
```

```python
#Michaelis Menten reaction
E = Symbol('E')
S = Symbol('S')
ES = Symbol('ES')
P = Symbol('P')
cme = CME()
cme.add_reaction(E + S, ES, E*S)
cme.add_reaction(ES, E + S, ES)
cme.add_reaction(ES, E + P, ES)

#for state, dt in cme.gillespie({S:10, E:2, ES:0, P:0}, 15):
#  print state, dt
```
