#La diffusione


la diffusione parte dalle leggi di Fick, che descrivono il flusso di concentrazione di una sostanza che si trovi in condizioni di non equilibrio (concentrazione non uniforme)

###prima legge di Fick

descrive la corrente di concentrazione dovuta alla derivata nello spazio della concentrazione stessa in termini di una costante chiamata coefficiente di diffusione.

$$J = -D \frac{\partial \phi}{\partial x}$$

oppure, in più dimensioni si può esprimere tramite il gradiente (la derivata in più dimensioni)

$$\vec J = -D \vec \nabla \phi$$

###seconda legge di Fick

Descrive il cambiamento di concentrazione nel tempo in funzione della sua derivata seconda, e può essere ricavata dalla prima legge sotto l'ipotesi di conservazione della massa.

$$\frac{\partial \phi}{\partial t} = D \frac{\partial^2 \phi}{\partial x^2}$$

Dal punto di vista analitico è importante notare che sono tutte ricavata consideranto il coefficiente di diffusività D costante e isotropo nello spazio e nel tempo ed indipendente dalla concentrazione attuale.

Le leggi della diffusione possono essere anche viste come l'effetto macroscopico di un enorme numero di particelle che si muovo in modo disordinato, compiendo quella che viene chiamata una random walk.

##random walk semplice 

la maniera più semplice per descrivere un random walk è quello immaginare una particella che si possa spostare in una sola dimensione, e che in ogni istante di tempo possa fare un salto in direzione positiva o negativa. Per visualizzare questo risultato possiamo estrarre un numero random e assegnare il valore +1 se il numero è sopra 0.5 e -1 se si trova al di sotto.

Usiamo un piccolo trucco in cui i valori True e False vengono convertiti in 1 e 0 nei calcoli numerici e possiamo ottenere con facilità il nostro step

```python
p = rand()
print p<0.5
print (p<0.5)*2 -1
```

per ottenere una random walk in maniera rapida posso sfruttare le capacità di numpy di creare e manipolare degli array con molti numeri in un colpo solo.

Se volessimo ad esempio simulare 10 passi:

```python
steps = (rand(10) <0.5) * 2 -1
print steps
```

Per ottenere la posizione dopo N passi dobbiamo sommare ciascuno step in successione. Possiamo farlo con un ciclo e salvando il risultato in un altro array, ma numpy ci viene in soccorso con la funzione **cumsum** che compie questo lavoro per noi

```python
print cumsum(steps)
```

```python
plot(cumsum(steps), linestyle='steps', linewidth=4)
```

ripeto per un maggior numero di step, ed osservo quello che viene definito un random walk, e che per un gran numero di step può essere approssimato ad un processo continuo

```python
steps = (rand(1000) <0.5) * 2 -1
plot(cumsum(steps), linestyle='steps', linewidth=2)
```

Possiamo in questa maniera simulare facilmente un gran numero di camminatori contemporaneamente

```python
N_steps = 100
N_walk = 1000
N = (N_walk, N_steps)
steps = np.random.randint(0.0,2.0,N)*2-1
walks = np.cumsum(steps, axis=1)
plot(walks.T, alpha=0.02, color='b', linestyle='steps');
```

----
###distribuzione binomiale

Mostriamo ora come la distribuzione della posizione dei camminatori con il tempo risulti essere una distribuzione binomiale (una volta normalizzata rispetto al minimo)

```python
walks_norm = (walks + arange(1, 1 + N_steps))/2.0
plot(walks_norm.T, alpha=0.02, color='b');
```

Confronto in un qualsiasi momento questa distribuzione con quella che mi aspetto di osservare da una distribuzione binomiale equivalente

```python
from scipy.stats import binom
pos = 100
plot(binom.pmf(range(pos), pos, 0.5), linestyle='steps-mid')
hist(walks_norm[:,pos-1], normed=True, bins=arange(100)-0.5, edgecolor='none');
plt.xlim(0,100)
```

###first passage time

Proviamo ora a considerare il problemadella fuga del camminatore dal piano, ovvero quando il suo valore scende sotto un valore prefissato. Possiamo ottenere quindi una distribuzione di quanti camminatori hanno raggiunto il minimo dopo un certo tempo.

Considero ad esempio quelli che scendono sotto zero partendo dalla posizione 10

```python
N_steps = 1000
N_walk = 1000
N = (N_walk, N_steps)
steps = (rand(*N)<0.5)*2-1
walks = 10 +np.cumsum(steps, axis=1)
```

```python
plot(walks.T, alpha=0.02, color='b', linestyle='steps');
```

vedo in ogni momento quanti sono sopra lo zero

```python
walks_over = walks > 0
```

Quando un camminatore ha toccato lo zero rimane zero per sempre

```python
walks_over = np.minimum.accumulate(walks_over, axis=1)
```

in ogni momento posso contare quanti sono i "sopravvissuti"

```python
survived = sum(walks_over, axis=0)
plot(survived)
gca().set_ylim(0.0, None)
```

##random walk in due dimensioni

posso simularlo semplicemente con due random walk indipendenti fra di loro


```python
N = 30
single_steps = (rand(N,2)<0.5)*2-1
single_walk = cumsum(single_steps, axis=0).T
plot(single_walk[0], single_walk[1], marker='o', linestyle='--', alpha=0.5)
```

```python
ora ne creo diversi tutti insieme e vedo che cosa succede
```

```python
N = 10000
NW = 100

radius = 2.0*sqrt(N)
single_steps = (rand(N, 2, NW)<0.5)*2-1
single_walk = cumsum(single_steps, axis=0).T
for walk in single_walk:
    plot(walk[0], walk[1], alpha=0.05, color='b')
plot([0],[0], marker='o', color='magenta')
gca().add_patch(Circle((0,0), radius, facecolor='none', edgecolor='r'))
plt.axes().set_aspect('equal', 'datalim')
show()
```

```python
from scipy.ndimage.filters import gaussian_filter
```

```python
initial = np.r_[0*np.ones(100), 1*np.ones(100), 0*np.ones(100), 1*np.ones(100), 0*np.ones(100)]
initial /= sum(initial)
results = [initial]
N = 60000
for i in range(N):
    res = gaussian_filter(results[-1], sigma=0.5)
    results.append(res/sum(res))
```

```python
for idx, i in enumerate(results):
    r= idx*1.0/N
    if not idx%200:
        plot(i, color=[1-r, 0.0, r])
```
