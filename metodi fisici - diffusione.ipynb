#La diffusione


la diffusione parte dalle leggi di Fick

###prima legge di Fick

sulle correnti

$$J = -D \frac{\partial \phi}{\partial x}$$

oppure, in più dimensioni

$$J = -D \nabla \phi$$

###seconda legge di Fick

la descrizione del cambiamento della concentrazione

$$\frac{\partial \phi}{\partial t} = D \frac{\partial^2 \phi}{\partial x^2}$$

può essere ricavata dalla prima legge sotto l'ipotesi di conservazione della massa
[<img src="http://upload.wikimedia.org/wikipedia/commons/thumb/4/4d/DiffusionMicroMacro.gif/250px-DiffusionMicroMacro.gif">](diagramma della diffusione)

```python
from scipy.ndimage.filters import gaussian_filter
```

```python
initial = np.r_[0*np.ones(100), 1*np.ones(100), 0*np.ones(100), 1*np.ones(100), 0*np.ones(100)]
initial /= sum(initial)
results = [initial]
N = 60000
for i in range(N):
    res = gaussian_filter(results[-1], sigma=0.5)
    results.append(res/sum(res))
```

```python
for idx, i in enumerate(results):
    r= idx*1.0/N
    if not idx%200:
        plot(i, color=[1-r, 0.0, r])
```

```python
initial = np.r_[0.0:1.0:1000j]
initial /= sum(initial)
results = [initial]
N = 60000
for i in range(N):
    res = gaussian_filter(results[-1], sigma=0.5)
    results.append(res/sum(res))
```

```python
for idx, i in enumerate(results):
    r= idx*1.0/N
    if not idx%200:
        plot(i, color=[1-r, 0.0, r])
```

```python
N_steps = 100
N_walk = 1000
N = (N_walk, N_steps)
steps = np.random.randint(0.0,2.0,N)*2-1
walks = np.cumsum(steps, axis=1)
plot(walks.T, alpha=0.02, color='b');
```

```python
walks_norm = (walks + arange(1, 1 + N_steps))/2.0
plot(walks_norm.T, alpha=0.02, color='b');
```
