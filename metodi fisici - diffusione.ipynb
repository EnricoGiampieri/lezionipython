#La diffusione


la diffusione parte dalle leggi di Fick, che descrivono il flusso di concentrazione di una sostanza che si trovi in condizioni di non equilibrio (concentrazione non uniforme)

###prima legge di Fick

descrive la corrente di concentrazione dovuta alla derivata nello spazio della concentrazione stessa in termini di una costante chiamata coefficiente di diffusione.

$$J = -D \frac{\partial \phi}{\partial x}$$

oppure, in più dimensioni si può esprimere tramite il gradiente (la derivata in più dimensioni)

$$\vec J = -D \vec \nabla \phi$$

###seconda legge di Fick

Descrive il cambiamento di concentrazione nel tempo in funzione della sua derivata seconda, e può essere ricavata dalla prima legge sotto l'ipotesi di conservazione della massa.

$$\frac{\partial \phi}{\partial t} = D \frac{\partial^2 \phi}{\partial x^2}$$

Dal punto di vista analitico è importante notare che sono tutte ricavata consideranto il coefficiente di diffusività D costante e isotropo nello spazio e nel tempo ed indipendente dalla concentrazione attuale.

Le leggi della diffusione possono essere anche viste come l'effetto macroscopico di un enorme numero di particelle che si muovo in modo disordinato, compiendo quella che viene chiamata una random walk.

Usaremo questa descrizione microscopica per ottenere informazioni statistiche a proposito della distribuzione macroscopica di probabilità.

##random walk semplice 

la maniera più semplice per descrivere un random walk è quello immaginare una particella che si possa spostare in una sola dimensione, e che in ogni istante di tempo possa fare un salto in direzione positiva o negativa. Per visualizzare questo risultato possiamo estrarre un numero random e assegnare il valore +1 se il numero è sopra 0.5 e -1 se si trova al di sotto.

Usiamo un piccolo trucco in cui i valori True e False vengono convertiti in 1 e 0 nei calcoli numerici e possiamo ottenere con facilità il nostro step

```python
p = rand()
print p<0.5
print (p<0.5)*2 -1
```

per ottenere una random walk in maniera rapida posso sfruttare le capacità di numpy di creare e manipolare degli array con molti numeri in un colpo solo.

Se volessimo ad esempio simulare 10 passi:

```python
steps = (rand(10) <0.5) * 2 -1
print steps
```

Per ottenere la posizione dopo N passi dobbiamo sommare ciascuno step in successione. Possiamo farlo con un ciclo e salvando il risultato in un altro array, ma numpy ci viene in soccorso con la funzione **cumsum** che compie questo lavoro per noi

```python
print cumsum(steps)
```

```python
plot(cumsum(steps), linestyle='steps', linewidth=4)
```

ripeto per un maggior numero di step, ed osservo quello che viene definito un random walk, e che per un gran numero di step può essere approssimato ad un processo continuo

```python
steps = (rand(1000) <0.5) * 2 -1
plot(cumsum(steps), linestyle='steps', linewidth=2)
```

Possiamo in questa maniera simulare facilmente un gran numero di camminatori contemporaneamente

```python
N_steps = 100
N_walk = 1000
N = (N_walk, N_steps)
steps = np.random.randint(0.0,2.0,N)*2-1
walks = np.cumsum(steps, axis=1)
plot(walks.T, alpha=0.02, color='b', linestyle='steps');
```

----
###distribuzione binomiale

Mostriamo ora come la distribuzione della posizione dei camminatori con il tempo risulti essere una distribuzione binomiale (una volta normalizzata rispetto al minimo)

```python
walks_norm = (walks + arange(1, 1 + N_steps))/2.0
plot(walks_norm.T, alpha=0.02, color='b');
```

Confronto in un qualsiasi momento questa distribuzione con quella che mi aspetto di osservare da una distribuzione binomiale equivalente

```python
from scipy.stats import binom
pos = 100
plot(binom.pmf(range(pos+1), pos, 0.5), linestyle='steps-mid')
hist(walks_norm[:,pos-1], normed=True, bins=arange(100)-0.5, edgecolor='none');
plt.xlim(0,100)
```

Potendo usare la formula della binomiale per descrivere la mia distribuzione nel tempo, posso ottenere con facilità una stima della deviazione standard al passaggio N:

$$\sigma = \frac{1}{2}\sqrt{N}$$

----
###first passage time

Proviamo ora a considerare il problemadella fuga del camminatore dal piano, ovvero quando il suo valore scende sotto un valore prefissato. Possiamo ottenere quindi una distribuzione di quanti camminatori hanno raggiunto il minimo dopo un certo tempo.

Considero ad esempio quelli che scendono sotto zero partendo dalla posizione 10

```python
N_steps = 1000
N_walk = 1000
N = (N_walk, N_steps)
steps = (rand(*N)<0.5)*2-1
walks = 10 +np.cumsum(steps, axis=1)
```

```python
plot(walks.T, alpha=0.02, color='b', linestyle='steps');
```

vedo in ogni momento quanti sono sopra lo zero

```python
walks_over = walks > 0
```

Quando un camminatore ha toccato lo zero rimane zero per sempre

```python
walks_over = np.minimum.accumulate(walks_over, axis=1)
```

in ogni momento posso contare quanti sono i "sopravvissuti"

```python
survived = sum(walks_over, axis=0)
plot(survived)
gca().set_ylim(0.0, None)
```

##random walk in due dimensioni

Posso prendere in considerazione anche dei random walk in più dimensioni in modo molto simile a quanto fatto nel caso monodimensionale. Per simulare ad esempio un random walk in due dimensioni posso semplicemente unire due random walk monodimensionali indipendenti, uno lungo la direzione x ed uno lungo la direzione y

```python
N = 30
single_steps = (rand(N,2)<0.5)*2-1

single_walk = cumsum(single_steps, axis=0).T

plot(single_walk[0], single_walk[1], marker='o', linestyle='--', alpha=0.5)
plt.axes().set_aspect('equal', 'datalim')
```

Ora ne creo diversi tutti insieme e ne disegno l'intera storia

```python
N = 10000
NW = 100

radius = sqrt(N)
single_steps = (rand(N, 2, NW)<0.5)*2-1
single_walk = cumsum(single_steps, axis=0).T
print single_walk.shape
```

```python
fig = figure()
ax = fig.add_subplot(1, 2, 1)
for walk in single_walk:
    ax.plot(walk[0], walk[1], alpha=0.05, color='b')
    
ax.plot([0],[0], marker='o', color='magenta')

ax.add_patch(Circle((0,0), radius, facecolor='none', edgecolor='r'))

ax.set_aspect('equal', 'datalim')

ax = fig.add_subplot(1, 2, 2)

ax.hist(single_walk[:,0,:].ravel(), normed=True, bins=50)
ax.axvline(100, color='r')
ax.axvline(-100, color='r')
fig.tight_layout()
show()
```

Posso anche visualizzare la posizione finale di tutti i vari camminatori, che mi dà la distribuzione di probabilità al tempo finale. Devo però usarne un maggior numero!

```python
N = 4000
NW = 10000

radius = sqrt(N)
print radius
single_steps = (rand(N, 2, NW)<0.5)*2-1
single_walk = cumsum(single_steps, axis=0).T
print single_walk.shape
```

###Integrazione al continuo

Dalla risoluzione delle equazioni differenziali possiamo vedere che nel caso di coefficiente di diffusione omogeneo e costante possiamo calcolare l'evoluzione nel tempo come la convoluzione della presente distribuzione di concentrazione con una gaussiana con deviazione standard pari a quella che ci attendiamo dal random walk nel tempo trascorso.

Questo è possibile in modo molto semplice usando la funzione gaussian_filter del modulo ndimage, che effettua in modo estremamente rapido la convoluzione con la funzione gaussiana

```python
from scipy.ndimage.filters import gaussian_filter
```

Per usarla ci basta fornire la distribuzione di concentrazione e la sigma della gaussiana che vogliamo utilizzare

```python
iniziale = np.r_[0*np.ones(100), 1*np.ones(100)]
plot(iniziale)
```

```python
successiva = gaussian_filter(iniziale, sigma=10)
plot(successiva)
```

L'applicazione successiva di questa funzione più volte ci permette di visualizzare l'andamento nel tempo di distribuzioni anche complicate

```python
initial = np.r_[0*np.ones(100), 1*np.ones(100), 0*np.ones(100), 1*np.ones(100), 0*np.ones(100)]
initial /= sum(initial)
plot(initial)
```

```python
N = 60000
results = [initial]
for i in range(N):
    res = gaussian_filter(results[-1], sigma=0.5)
    results.append(res/sum(res))
```

```python
for idx, i in enumerate(results):
    r= idx*1.0/N
    if not idx%200:
        plot(i, color=[1-r, 0.0, r])
```

Con un pò di fantasia possiamo usare questa funzione per simulare la diffusione in mezzi a diffusività non omogenea. 

Consideriamo un problema monodimensionale, con concentrazione uniforme ed una zona centrale con diffusività minore del resto (ad esempio una gelatina immersa in acqua).

Possiamo simulare questo sistema con due step simili ai precedenti: Simulo prima l'evoluzione della zona a diffusività maggiore (ignorando quella a diffusività minore), simulo la seconda zona ed unisco i risultati.

Inizio creando la distribuzione iniziale ed un array che mi permetta di distinguere le due zone

```python
iniziale = ones(1000)

zone = r_[ ones(450), zeros(100), ones(450) ]==True
```

Ora faccio la diffusione considerendo diverse da zero solo le aree interessate con la funzione **where**, che dato un vettore di booleani selezione i valori fra due array o scalari

```python
where( [ 1, 1, 0 ], 14, 22 )
```

```python
concentratrazione_1 = where( zone, iniziale, 0.0 )
evoluzione_1 = gaussian_filter( concentratrazione_1, 5.0 )

concentratrazione_2 = where( zone, 0.0, iniziale )
evoluzione_2 = gaussian_filter( concentratrazione_2, 1.0 )

nuova_concentrazione = evoluzione_1 + evoluzione_2
plot(nuova_concentrazione)
```

Ora simulo molti steps per determinare il risultato

```python
N = 1000
results = [iniziale / sum(iniziale) ]
for i in range(N):
    concentratrazione_1 = where( zone, results[-1], 0.0 )
    evoluzione_1 = gaussian_filter( concentratrazione_1, 5.0 )

    concentratrazione_2 = where( zone, 0.0, results[-1] )
    evoluzione_2 = gaussian_filter( concentratrazione_2, 1.0 )

    nuova_concentrazione = evoluzione_1 + evoluzione_2
    results.append(nuova_concentrazione / sum(nuova_concentrazione))
    
for idx, i in enumerate(results):
    r= idx*1.0/N
    if not idx%(N//20):
        plot(i, color=[1-r, 0.0, r])
```

Possiamo vedere come nel tempo la zona a diffusività minore tende ad assorbire buona parte della concentrazione. Questo ha un effetto profondo sulla diffusione all'interno del mezzo cellulare.
