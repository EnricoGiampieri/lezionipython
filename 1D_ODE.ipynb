##Integration of a system of two differential equation 

$\displaystyle dy_0(t)dt=f_0(y_0,y_1,t)$  

$\displaystyle dy_1(t)dt=f_2(y_0,y_1,t)$  

##y(t)=odeint(f(y,t),y0,time)

# $\dot C_o = +k_{co} C_c - k_{oc} C_o$

# $C_o =0, C_c = N$

#$ C_o = N \frac{k_{oc}}{k_{oc}+k_{co}} $
#$ C_c = N - C_o $

```python
import numpy
from scipy import integrate
import pylab



def f(y, t): 
    
  R = y[0] 
  P = y[1] 
  eq1=-a*R+b*P
  eq2= a*R-b*P  
    
  return [eq1,eq2]

  
  
  #Parameters assignements
a=2
b=1  
    #initial conditions
R0=1
P0=0
y0=[R0,P0]
    
t = arange(0, 5, 0.1)
    
r = integrate.odeint(f, y0, t)

ylim(0,1.5)

plot(t, r, '-')
  #pylab.legend(['$C_R(t)$', '$C_P(t)$'], loc='upper right',prop={'size':15})
text(4,0.75, '$C_P(eq)$',size=18)
text(4,1.1, '$C_T$',size=18)
text(4,0.40, '$C_R(eq)$',size=18)
xlabel('Time',size=18)
ylabel('Concentration',size=18)
title('First order reversible reaction',size=18)
plot([0,5], [1,1], 'k--', lw=2)
  
show()
```

# Bimolecular reaction:  

 
$ A+B{{k_{f} \atop\displaystyle\longrightarrow} \atop {\displaystyle\longleftarrow \atop
 k_{b}}}C $  

$
\begin{cases}
\displaystyle
 \frac{d C_A}{dt}=-k_f\cdot C_A\cdot C_B+k_b\cdot C_C \\\\
 \quad\\\\
 \displaystyle
 \frac{d C_B}{dt}=-k_f\cdot C_A\cdot C_B+k_f\cdot C_C\\\\
 \quad\\\\
  \displaystyle
 \frac{d C_C}{dt}=+k_f\cdot C_A\cdot C_B-k_b\cdot C_C\\\\
\end{cases}
$


```python
import numpy
from scipy import integrate
import pylab



def f(y, t): 
    
  A = y[0] 
  B = y[1] 
  C = y[2]
  eq1=-kf*A*B+kb*C
  eq2= -kf*A*B+kb*C
  eq3= +kf*A*B-kb*C 
    
  return [eq1,eq2,eq3]

  
  
  #Parameters assignements
kf=10.
kb=1.  
    #initial conditions
A0=1.
B0=.5
C0=0
y0=[A0,B0,C0]
    
t = arange(0, 5, 0.1)
    
r = integrate.odeint(f, y0, t)

ylim(0,2) #setting the y limit

plot(t, r, '-') #'-' is for continuos line

legend(['$A(t)$', '$B(t)$','$C(t)$'], loc='upper right',prop={'size':10})
#text(4,0.75, '$C_P(eq)$',size=18)
#text(4,1.1, '$C_T$',size=18)
text(1,1.6, '$A+B+2C$',size=10)
xlabel('Time',size=18)
ylabel('Concentration',size=18)
title('Bimolecular reaction',size=18)
plot([0,5], [1.5,1.5], 'k--', lw=2)#plotting an horizontal line
#savefig("/home/gastone/Me/Did/Systems Biology_2013/Figures_1/Bimolecular.png",format='png') 
show()

s=r[:,[0]]+r[:,[1]]+2*r[:,[2]]#check of the conservation of mass
plot(t,s)
show()
```

#Enzyme kinetic: Michaelis Menten reaction:  

#$E+S{{k_{f} \atop\longrightarrow} \atop {\longleftarrow \atop  k_{b}}}ES{{k_{c}\atop\longrightarrow} \atop {\quad  \atop \quad }}E+P$
 

\begin{eqnarray}
\begin{cases}
{d E(t)\over d t}&=& -k_f \cdot  E\cdot S + k_b\cdot ES + k_c\cdot ES\\\\
\\\\
{d S(t)\over d t}&=& -k_f \cdot  E\cdot S + k_b\cdot ES\\\\ 
\\\\
{d ES(t)\over d t}&=& k_f \cdot  E\cdot S - k_b\cdot ES - k_c\cdot ES\\\\
\\\\
{d P(t)\over d t}&=& k_c\cdot ES\\
\end{cases}
\end{eqnarray}  







```python
import numpy
from scipy import integrate
import pylab



def f(y, t): #function definition
    
  E = y[0] 
  S = y[1] 
  ES = y[2]
  P = y[3]
  eq1=-kf*E*S+kb*ES+kc*ES
  eq2=-kf*E*S+kb*ES
  eq3= kf*E*S-kb*ES-kc*ES
  eq4=kc*ES
    
  return [eq1,eq2,eq3,eq4]

  
  
  #Parameters assignements
kf=1.
kb=0.1
kc=1.
km=(kb+kc)/kf
    #initial conditions
E0=1
S0=1000
ES0=0
P0=0
y0=[E0,S0,ES0,P0]
    
t = arange(0, 2000, 0.1)
    
r = integrate.odeint(f, y0, t)

ylim(0,S0+1)

plot(t, r, '-')

legend(['$E(t)$', '$S(t)$','$ES(t)$','$P(t)$'], loc='upper right',prop={'size':10})
xlabel('Time',size=18)
ylabel('Concentrations',size=18)
title('Michaelis Menten reaction',size=18)
#plot([0,5], [1,1], 'k--', lw=2)
#savefig("/home/gastone/Me/Did/Systems Biology_2013/Figures_1/MM_full.png",format='png') 
show()

#plot(sort(r[:,[1]]),sort(r[:,[2]]))



```

#Comparison of the steady state solution and the full solution  
By appliyng the quasi steady state approximation, the MM system  

is reduced to two equations   

$
\begin{eqnarray}
\begin{cases}
{d S(t)\over d t}&=& -k_c \cdot  \frac{E_T\cdot S}{k_m+S}\\\\ 
\\\\
{d P(t)\over d t}&=& +k_c \cdot  \frac{E_T\cdot S}{k_m+S}\\
\end{cases}
\end{eqnarray}   
$  

The Michaelis–Menten equilibrium analysis is valid if the substrate reaches equilibrium on a much faster time-scale than the product is formed:

$\displaystyle \epsilon_d = \frac{k_c}{k_b}  $ <<1

```python
import numpy
from scipy import integrate
import pylab

  #Parameters assignements
kf=1.
kb=0.1
kc=1.
km=(kb+kc)/kf

    #initial conditions
E0=0
S0=15
ES0=1
P0=0
y0=[E0,S0,ES0,P0]
z0=[S0,P0]

ET=E0+ES0
def f(y, t): 
  E = y[0] 
  S = y[1] 
  ES = y[2]
  P = y[3]
  eq1=-kf*E*S+kb*ES+kc*ES
  eq2=-kf*E*S+kb*ES
  eq3= kf*E*S-kb*ES-kc*ES
  eq4=kc*ES
  return [eq1,eq2,eq3,eq4]

def g(z, t): 
    S=z[0]
    P=z[1]
    eq1=-kc*(ET)*S/(km+S)
    eq2=kc*(ET)*S/(km+S)
    return [eq1,eq2]
  

t = arange(0, 50, 0.1)
    
r = integrate.odeint(f, y0, t)
s = integrate.odeint(g, z0, t)

ylim(0,16)


plot(t, r, '-')
legend(['$E(t)$', '$S(t)$','$ES(t)$','$P(t)$'], loc='upper right',prop={'size':10})
plot(t,s,'--')
xlabel('Time',size=18)
ylabel('Concentration',size=18)
title('Michaelis Menten reaction',size=18)
#plot([0,5], [1,1], 'k--', lw=2)
#savefig("/home/gastone/Me/Did/Systems Biology_2013/Figures_1/MM_comp_kc_1_kb_20.png",format='png')  
show()

plot(t,s)
show()
#plot(sort(r[:,[1]]),sort(r[:,[2]]))
```

#The Hill function  
#$\displaystyle \theta = {[L]^n \over K_d + [L]^n} = {[L]^n \over (K_A)^n + [L]^n}$  

This function is used in all the situations where we have a "cooperativity effect".  
This effect is quantified by the parameter $n$.  

$\begin{cases}
n<1\quad \text{negative cooperativity, other ligands have a unfavorable effect   }\\\\
n=1 \quad \text{no cooperativity, independence from other ligands}\\\\
n>1 \quad \text{positive cooperativity, other ligands have a favorable effect   }\\
\end{cases}
$

```python
import numpy
from scipy import integrate
import pylab

l=arange(0,20,0.1)
vm=1.
def hill(n,k):
    temp=vm*l**n/(k**n+l**n)
    return temp



fig=plt.figure(figsize=(15, 4))
fig.suptitle('Hill plots', fontsize=20,y=1.1)
#plt.figure(1)
plt.subplot(1,3,1)
ylim(0,1.5)#setting the y limit
xlabel('$L$ concentration',size=18)
ylabel('Rate :' '$v(L) $',size=18)
title('$n=.2$,$K_d$=2',size=18)
plot( l,hill(.2,5) )
#plt.show()
plt.subplot(1,3,2)
ylim(0,1.5)#setting the y limit
xlabel('$L$ concentration',size=18)
ylabel('Rate :' '$v(L) $',size=18)
title('$n=1$,$K_d$=2',size=18)
plot(l,hill(1.0,5))
plt.subplot(1,3,3)
ylim(0,1.5)#setting the y limit
xlabel('$L$ concentration',size=18)
ylabel('Rate :' '$v(L) $',size=18)
title('$n=5$,$K_d$=2',size=18)
plot(l,hill(5,5))
#savefig("/home/gastone/Me/Did/Systems Biology_2013/Figures_1/Hill_K_d_fixed_less_1.png",format='png') 
plt.show()
```

#The Lotka-Volterra equations  

$
\begin{cases}
\displaystyle \frac{dx}{dt} = \alpha\cdot x - \beta\cdot x\cdot y\\\\
\quad \\\\
\displaystyle \frac{dy}{dt} = \delta\cdot x\cdot y-\gamma \cdot y\\

$

#Competitive Lotka Volterra model

$
\begin{cases}
\displaystyle
{dx_1 \over dt} = r_1x_1-\alpha_1\cdot x_1^2-\beta_1\cdot x_1\cdot x_2 \\\\
\quad\\\\
\displaystyle
{dx_2 \over dt} = r_1x_2-\alpha_1\cdot x_2^2-\beta_2\cdot x_2\cdot x_1.\\
\end{cases}
$

```python
import numpy
from scipy import integrate
import pylab



def f(y, t): 
    
  x1 = y[0] 
  x2 = y[1] 
  
  eq1=r1*x1-a1*x1**2-b1*x1*x2
  eq2=r2*x1-a2*x2**2-b2*x2*x1 
    
  return [eq1,eq2]

  
  
  #Parameters assignements
    
r1=5.
a1=1.
b1=2.
    
r2=3.
a2=1.
b2=0.5

    #initial conditions- to change to see the effect
x10=.1
x20=.2
y0=[x10,x20]
    
t = arange(0, 20, 0.1)
    
r = integrate.odeint(f, y0, t)

plot(t, r, '-') #'-' is for continuos line
ylim(0,max(max(r[:,[0]]),max(r[:,[1]]))+.5) #setting the y limit
legend(['$x_1(t)$', '$x_2(t)$'],loc='upper right',prop={'size':12})
xlabel('Time',size=18)
ylabel('Population $x_1(t)$, $x_2(t)$',size=18)
title('Competitive Lotka Volterra model',size=18)
#savefig("/home/gastone/Me/Did/Systems Biology_2013/Figures_1/LV_1112.png",format='png') 
show()
max(max(r[:,[0]]),max(r[:,[1]]))
```

### a) Malthusian process  

If $N$ is indicating the number of bacteria, their replication rate (Velocity of replication) is proportional to that number:
 

 
# $\displaystyle \frac{dN}{dt}=\alpha\cdot N$   
    
 Malthus principle (exponential growth in an unlimited medium).  
    
The solution is $\displaystyle N(t)=N(0)\cdot e^{\alpha\cdot t}$
    


```python
from scipy import integrate 
import pylab 

def f(y,t): # f(x) Function
     k=1
     return k*y
  
y0=50

time = linspace(0.0,5.0,1000)
y = integrate.odeint(f,y0,time)
ylim(0,500)
plot(time,y)
xlabel('time (arbitrary units)',size=20)
ylabel('nunber of bacteria $N(t)$', size =20) 
title('Malthusian growth', size =20)
annotate("$N(0)$",fontsize=20, xy=(0, 50), xytext=(.1, 130),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
#savefig("/home/gastone/Me/Did/Systems Biology_2013/Figures_1/Malthus.png",format='png')
show() 
```

### b) Malthusian process with constant Birth and Death rate  
A refinement of the model is based on the definition of Birth rate $B$ and mortality rate $D$:  
    
 $\displaystyle \frac{dN}{dt}=(B-D)\cdot N$  
    
### Obs  
    
    
B-D>0  

B-D=0  

B-D<0  



```python
from scipy import integrate 
import pylab 

def f(y,t): # f(x) Function
     B=1
     D=2
     return (B-D)*y
  
y0=50

time = linspace(0.0,5.0,1000)
y = integrate.odeint(f,y0,time)
ylim(0,500)
plot(time,y)
xlabel('time (arbitrary units)',size=20)
ylabel('nunber of bacteria $N(t)$', size =20) 
title('Malthusian growth', size =20)
annotate("$N(0)$",fontsize=20, xy=(0, 50), xytext=(.1, 130),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )
#savefig("/home/gastone/Me/Did/Systems Biology_2013/Figures_1/Malthus_BD.png",format='png')
show() 
```

### c) Logistic Growth  

 Experimentally we observe that the growth of a population is not unlimited, but it reach a saturation level
 The limiting factor is modeled as $N^2$ (depends from  the density of the colony)  

 

  replication term: $\alpha\cdot N$  

  limiting term: $\gamma\cdot N^2$  
  
  # $\displaystyle \frac{dN}{dt}=\alpha\cdot N-\gamma\cdot N^2$  
 
   
 The term in $N^2$ is indicating the effect due to the population size (the limiting factor).

```python
from scipy import integrate 
import pylab 

def f(y,t): # f(x) Function
     a=50
     d=2
     return a*y-d*y**2
  
y0=1

time = linspace(0.0,0.2,1000)
y = integrate.odeint(f,y0,time)
ylim(0,50)
plot(time,y)
xlabel('time (arbitrary units)',size=20)
ylabel('nunber of bacteria $N(t)$', size =20) 
title('Logistic growth', size =20)
#annotate("$N(0)$",fontsize=20, xy=(0, 50), xytext=(.1, 130),
 #           arrowprops=dict(facecolor='black', shrink=0.05),
  #          )
#savefig("/home/gastone/Me/Did/Systems Biology_2013/Figures_1/Logistic.png",format='png')
show() 
```

#Bistable model  

$
\begin{cases}
\displaystyle
\frac{dx_1}{dt}=x_1^2-x_1\cdot(x_1^2+x_2^2)\\\\
\quad\\\\
\displaystyle
\frac{dx_2}{dt}=x_2^2-x_2\cdot(x_1^2+x_2^2)\\\\
\end{cases}
$

```python
import numpy
from scipy import integrate
import pylab



def f(y, t): 
    
  x1 = y[0] 
  x2 = y[1] 
  
  eq1=x1**2-x1*(x1**2+x2**2)
  eq2=x2**2-x2*(x1**2+x2**2) 
    
  return [eq1,eq2]
#initial conditions- to change to see the effect
x10=.2
x20=.19999
y0=[x10,x20]
    
t = arange(0, 100, 0.1)
    
r = integrate.odeint(f, y0, t)

plot(t, r, '-') #'-' is for continuos line
ylim(0,max(max(r[:,[0]]),max(r[:,[1]]))+.5) #setting the y limit
legend(['$x_1(t)$', '$x_2(t)$'],loc='upper right',prop={'size':12})
xlabel('Time',size=18)
ylabel('Population $x_1(t)$, $x_2(t)$',size=18)
title('Competitive model',size=18)
#savefig("/home/gastone/Me/Did/Systems Biology_2013/Figures_1/LV_1112.png",format='png') 
show()
print(float(r[999,[0]]))
print(float(r[999,[1]]))


```

#The toggle switch  
$
\begin{cases}
\displaystyle
\frac{du}{dt}=\frac{\alpha_1}{1+v^{\beta}}-u\\\\
\quad\\\\
\displaystyle
\frac{dv}{dt}=\frac{\alpha_2}{1+u^{\gamma}}-v\\
