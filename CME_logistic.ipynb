```python
%load_ext sympy.interactive.ipythonprinting
from CMElib_new import CME
from sympy import Symbol
import math
```

```python
#logistic growth
A= Symbol('A')
cme = CME()
cme.add_reaction(A, None, .01*A**2)
cme.add_reaction(None, A, A)
```

```python
#visualizing the  CME
display(cme.writeCME())
```
